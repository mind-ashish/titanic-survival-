```python
#data cleaning in both train & test
```

```python
for i in range(0,668,1):
    if x_train.at[i,'Embarked']=='C':
        x_train.at[i,'Embarked']=0
    elif x_train.at[i,'Embarked']=='S':
        x_train.at[i,'Embarked']=1
    else:
        x_train.at[i,'Embarked']=2
```

```python
for i in range(0,223,1):
    if x_test.at[i,'Embarked']=='C':
        x_test.at[i,'Embarked']=0
    elif x_test.at[i,'Embarked']=='S':
        x_test.at[i,'Embarked']=1
    else:
        x_test.at[i,'Embarked']=2
```

```python
for i in range(0,668,1):
    if x_train.at[i,'Sex']=='male':
        x_train.at[i,'Sex']=0
    else:
        x_train.at[i,'Sex']=1
```

```python
for i in range(0,223,1):
    if x_test.at[i,'Sex']=='male':
        x_test.at[i,'Sex']=0
    else:
        x_test.at[i,'Sex']=1
```

```python
import math
sumAgeSurvived,sumAgeUnsurvive,countSurvived,countUnsurvived=0,0,0,0
for i in range(0,668,1):
    if not(math.isnan(x_train.at[i,'Age'])):
        if x_train.at[i,'Survived']==1:
            sumAgeSurvived+=x_train.at[i,'Age']
            countSurvived+=1
        else:
            sumAgeUnsurvive+=x_train.at[i,'Age']
            countUnsurvived+=1
    else:
        continue
        
meanSurvivedAge=sumAgeSurvived/countSurvived
meanUnsurvivedAge=sumAgeUnsurvive/countUnsurvived
meanAge=(sumAgeSurvived+sumAgeUnsurvive)/(countSurvived+countUnsurvived)
for i in range(0,668,1):
    if math.isnan(x_train.at[i,'Age']):
        if x_train.at[i,'Survived']==1:
            x_train.at[i,'Age']=meanSurvivedAge
        else:
            x_train.at[i,'Age']=meanUnsurvivedAge
    else:
        continue
count,age=0,0
for i in range(0,223,1):
    if not(math.isnan(x_test.at[i,'Age'])):
        age+= x_test.at[i,'Age']
        count+=1
meanAge=age/count
for i in range(0,223,1):
    if math.isnan(x_test.at[i,'Age']):
        x_test.at[i,'Age']=meanAge
```
